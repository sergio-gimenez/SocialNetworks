```python
import twitter
from twitter import *
CONSUMER_KEY = ""
CONSUMER_SECRET = ""
OAUTH_TOKEN = ""
OAUTH_TOKEN_SECRET = ""


auth = OAuth(
    consumer_key=CONSUMER_KEY,
    consumer_secret=CONSUMER_SECRET,
    token=OAUTH_TOKEN,
    token_secret=OAUTH_TOKEN_SECRET
)
t = Twitter(auth=auth)
print(t)
```

## World trends
Structure of the provided information. List (of one element) of dictionaries


```python

WORLD_WOE_ID = 1


world_trends = t.trends.place(_id=WORLD_WOE_ID,count=500)

# Transform an iterable into a list of dict
Results = list(world_trends)
print(world_trends)
```

### Contents of the results

```python
Results = list(world_trends)[0]
print(f'Number of results: {len(Results)}')
print(Results.keys())
print('Number of results: {}'.format(len(Results['trends'])))
```

##  <font color=blue>  Trends into a pandas data frame </font>
Data frame will be the main data structure (in this subject) for analysing the data from twitter.

```python
import pandas as pd
trends_df = pd.DataFrame.from_dict(Results['trends'])
trends_df
```

### Structure of the data frame

```python
trends_df.columns
```

### Sort by 'tweet_volume', reorder columns

```python
trends_df.sort_values(by = ['tweet_volume'], ascending=False)[['tweet_volume','name','query']]
```

### Lets add a new column and sort by several criteria

```python
trends_df['Length name'] = trends_df['name'].apply(lambda x: len(x))
trends_df[['tweet_volume','name','Length name']]
```

### sort by several criterion, and save in place.

```python
trends_df.sort_values(by = ['tweet_volume','Length name'], ascending = False, inplace = True)
trends_df[['tweet_volume','Length name']]
```
